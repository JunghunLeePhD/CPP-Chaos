# Dynamics of Logistic map

## The Logistic Map and its Dynamics

> **Definition (Logistic Map)**
> Let $\lambda > 0$ be a real parameter. We define the **logistic map** $\mathrm{L}_{\lambda}: \mathbb{R} \to \mathbb{R}$ by
>
> $$
> \mathrm{L}_{\lambda}(x) := \lambda x (1-x)
> $$
>
> for each $x \in \mathbb{R}$.

We consider the discrete dynamics generated by this map. That is, for a given initial condition $x_0 \in \mathbb{R}$, we wish to compute its orbit (trajectory) defined by the iteration:

$$
x_0 \xrightarrow{\mathrm{L}_{\lambda}} x_1 = \mathrm{L}_{\lambda}(x_0) \xrightarrow{\mathrm{L}_{\lambda}} x_2 = \mathrm{L}_{\lambda}(x_1) = \mathrm{L}_{\lambda}(\mathrm{L}_{\lambda}(x_0)) \xrightarrow{\mathrm{L}_{\lambda}} \cdots
$$

Explicitly, the second iterate is given by:

$$
x_2 = \lambda (\lambda x_0 (1 - x_0)) (1 - \lambda x_0 (1 - x_0)).
$$

While it is always possible to numerically compute the orbit for a specific initial condition $x_0$ and parameter $\lambda$, the qualitative behavior of the dynamics changes drastically with $\lambda$.

We first characterize the trivial global dynamics, identifying the region where orbits diverge.

> **Lemma 1**
> Let $\lambda > 1$. The logistic map admits fixed points at $0$ and $p_{\lambda} = 1 - \lambda^{-1}$. Moreover, orbits starting outside the unit interval diverge:
> 1. If $x < 0$, then $\mathrm{L}_{\lambda}(x) < \lambda x$ (since $1-x > 1$), and the orbit diverges monotonically to $-\infty$.
> 2. If $x > 1$, then $\mathrm{L}_{\lambda}(x) < 0$, mapping the point to the negative domain, where it subsequently diverges to $-\infty$.

A phenomenon of particular interest is that for generic initial conditions $x_0 \in (0, 1)$, we may lack information about the long-term orbit behavior, even if the orbit of $x_0 = 0$ is known. This is a typical behavior of chaotic dynamics, known as **sensitivity to initial conditions**. Roughly speaking, this implies that the system does not allow us to predict the 'future' state $\mathrm{L}_{\lambda}^n(y)$ based on the orbit of a nearby point $x$, because the error grows exponentially.

> **Definition (Sensitivity to Initial Conditions)**
> Let $f : X \to X$ be a map. A map $f: X \to X$ is said to have **sensitivity to initial conditions** if there exists a $\delta > 0$ (the sensitivity constant) such that for any $x \in X$ and any neighborhood $N$ of $x$, there exists a point $y \in N$ and a non-negative integer $n$ such that:
>
> $$
> |\mathrm{L}^n(x) - \mathrm{L}^n(y)| > \delta.
> $$

As Lemma 1 shows, the non-trivial dynamics of the logistic map are confined to the invariant closed unit interval $[0, 1]$. It is within this bounded region that chaotic dynamics appear.

## Logistic Dynamics I: The Stable Regime

In this subsection, we analyze the dynamics of the logistic map with the parameter $\lambda$ satisfying $1 < \lambda < 3$. In this regime, the system exhibits stable behavior, dominated by a single non-trivial attractor.

> **Lemma 2**
> Let $\lambda$ be a real number such that $1 < \lambda < 3$. Then, for any initial condition $x \in (0, 1)$, the orbit converges to the non-trivial fixed point:
>
> $$
> \lim_{n \to \infty} \mathrm{L}_{\lambda}^n(x) = p_\lambda = 1 - \frac{1}{\lambda}.
> $$

Combining this result with Lemma 1, we obtain a complete characterization of the global dynamics for this parameter range:

* **Divergence:** If $x < 0$ or $x > 1$, the orbit diverges to negative infinity:
    $$
    \lim_{n \to \infty} \mathrm{L}_{\lambda}^n(x) = -\infty.
    $$
* **Extinction:** If $x = 0$ or $x = 1$, the orbit eventually settles at the origin:
    $$
    \lim_{n \to \infty} \mathrm{L}_{\lambda}^n(x) = 0.
    $$
    (Note: $\mathrm{L}_{\lambda}(1) = 0$, so $x=1$ maps immediately to the fixed point at the origin).
* **Convergence to Equilibrium:** If $x \in (0, 1)$, the orbit asymptotically approaches the carrying capacity:
    $$
    \lim_{n \to \infty} \mathrm{L}_{\lambda}^n(x) = 1 - \frac{1}{\lambda}.
    $$

Thus, we have completely characterized the asymptotic behavior of the system for $1 < \lambda < 3$.

## Attracting Cycles and Basins of Attraction

As we observed in the previous regime ($1 < \lambda < 3$), the dynamics were dominated by a single attracting fixed point. To understand behavior for larger $\lambda$, we must generalize this concept to periodic cycles.

> **Definition (Periodic Point)**
> Let $f: \mathbb{R} \to \mathbb{R}$ be a map. A point $p$ is called a **periodic point** of period $k$ if $f^k(p) = p$ and $f^i(p) \neq p$ for all $1 \le i < k$. The sequence $\{p, f(p), \dots, f^{k-1}(p)\}$ is called a **periodic orbit** (or cycle) of period $k$.

Periodic points not only determine their own orbits but also dictate the behavior of nearby points.

> **Definition (Attracting Periodic Point)**
> A periodic point $p$ of period $k$ is said to be **attracting** (or linearly stable) if there exists an open neighborhood $U$ of $p$ such that for all $x \in U$, the orbit of $x$ under the map $f^k$ converges to $p$:
>
> $$
> \lim_{n \to \infty} (f^k)^n(x) = p.
> $$

Determining the existence of such a neighborhood directly from the definition can be difficult. Fortunately, linear stability analysis provides a practical calculation based on the derivative along the orbit.

> **Proposition**
> Let $\{x_0, x_1, \dots, x_{k-1}\}$ be a periodic orbit of period $k$. This orbit is attracting if and only if the absolute value of the derivative of the $k$-th iterate at any point $x_i$ in the orbit is less than 1. By the Chain Rule, this is equivalent to:
>
> $$
> |(f^k)'(x_0)| = \left|\prod_{i=0}^{k-1} f'(x_i)\right| < 1.
> $$

The set of all points whose orbits converge to an attracting periodic cycle is open. We formally define this region as the basin of attraction.

> **Definition (Basin of Attraction)**
> Let $\mathcal{O}$ be an attracting periodic orbit. The **basin of attraction** of $\mathcal{O}$, denoted $\mathcal{B}(\mathcal{O})$, is the set of all initial conditions $x \in \mathbb{R}$ such that the orbit of $x$ converges to $\mathcal{O}$ as $n \to \infty$. The connected component of $\mathcal{B}(\mathcal{O})$ containing the attractor itself is called the **immediate basin of attraction**.

Note that for the logistic map, the basin of attraction cannot be the entire real line $\mathbb{R}$. For instance, points $x > 1$ map to negatives and diverge to $-\infty$. Additionally, boundaries exist between the basins of different attractors (e.g., the basin of the periodic cycle vs. the basin of infinity). Points on this boundary exhibit sensitivity to initial conditions.

Crucially, a powerful result known as **Singer's Theorem** states that for maps with a negative Schwarzian derivative (such as the logistic map), the immediate basin of any attracting periodic orbit must contain a critical point of the map. Since the logistic map has only one critical point ($c = 1/2$), we obtain a profound restriction on its dynamics:

> **Theorem**
> The logistic map $\mathrm{L}_{\lambda}$ has at most one attracting periodic orbit. If such an orbit exists, its basin of attraction must contain the critical orbit starting at $x = 1/2$.

Consequently, we can determine the global asymptotic behavior of the system simply by observing the orbit of the critical point $x = 1/2$.

## Logistic Dynamics II: Symbolic Dynamics

We now turn our attention to the regime where $\lambda > 4$. In this case, the dynamics change fundamentally because the interval $[0, 1]$ is no longer invariant. The maximum value of the map, $\mathrm{L}_{\lambda}(1/2) = \lambda/4$, strictly exceeds $1$. Consequently, the critical point maps to the region $x > 1$, and subsequently diverges to $-\infty$.

It can be shown that for almost all initial conditions $x_0 \in [0, 1]$, the orbit escapes to $-\infty$. The set of points that \textit{never} escape forms a fractal subset of the interval known as a **Cantor set**, denoted by $\Lambda$. On this invariant set $\Lambda$, the dynamics are chaotic. In fact, all periodic points within $\Lambda$ are repelling (hyperbolic), and the set contains infinitely many such points.

To rigorously analyze the chaos in this regime, we employ **symbolic dynamics**.

> **Definition**
> Let $\Sigma_2 = \{0, 1\}^{\mathbb{N}}$ be the space of all infinite sequences of two symbols, $s = (s_0, s_1, \dots)$, where $s_i \in \{0, 1\}$. We define the **shift map** $\sigma: \Sigma_2 \to \Sigma_2$ by:
>
> $$
> \sigma(s_0, s_1, s_2, \dots) = (s_1, s_2, s_3, \dots).
> $$
>
> The space $\Sigma_2$ is equipped with a metric
>
> $$
> d(s, t) = \sum_{i=0}^{\infty} \frac{|s_i - t_i|}{2^i},
> $$
>
> under which $\sigma$ is continuous. The shift map has sensitivity to initial conditions, topological mixing, and dense periodic orbits.

The power of symbolic dynamics lies in its ability to completely model the complex trajectory of the logistic map through a topological equivalence.

> **Proposition**
> The logistic map $\mathrm{L}_{\lambda}$ restricted to the invariant Cantor set $\Lambda$ is topologically conjugate to the shift map on two symbols. That is, there exists a homeomorphism $h: \Lambda \to \Sigma_2$ such that the following diagram commutes:
>
> $$
> h \circ \mathrm{L}_{\lambda} = \sigma \circ h.
> $$

**Remark on Metrics and Expansion:**
The existence of this homeomorphism relies on the map being \textit{uniformly expanding} on the invariant set $\Lambda$. The choice of metric required to prove this expansion depends on the magnitude of $\lambda$:

1.  **Large Parameter ($\lambda > 2+\sqrt{5}$):** If $\lambda$ is large enough, specifically $\lambda > 2+\sqrt{5} \approx 4.236$, the derivative satisfies $|\mathrm{L}'_{\lambda}(x)| > 1$ for all $x$ in the intervals containing $\Lambda$. In this case, the expansion can be proven directly using the standard **Euclidean metric**.
2.  **Parameter close to 4 ($4 < \lambda \le 2+\sqrt{5}$):** In this regime, the derivative $|\mathrm{L}'_{\lambda}(x)|$ may be less than $1$ in certain regions of the occupied intervals. To prove expansion here, one cannot rely on the Euclidean metric. Instead, one must view the map as a complex dynamical system and utilize the **hyperbolic metric** (PoincarÃ© metric) on the complement of the escaped set in the complex plane. By the Schwarz-Pick Lemma, the map expands distances in this hyperbolic metric, ensuring the conjugacy holds.

```python
import matplotlib.pyplot as plt
import numpy as np

def logistic_map(x, lam):
    """
    The logistic map function.
    f(x) = lambda * x * (1 - x)
    """
    return lam * x * (1 - x)

def get_cobweb_path(lam, start_val, n_iter):
    """
    Helper function to calculate the coordinates for a single cobweb trajectory.
    Returns tuple (px, py) containing the x and y coordinates of the path.
    """
    x = start_val
    px, py = [], []

    # Start at (start_val, 0) to move up to the curve
    px.append(x)
    py.append(0)

    for _ in range(n_iter):
        x_next = logistic_map(x, lam)

        # Vertical line: move from (x, x) or (x, 0) up to curve (x, f(x))
        px.append(x)
        py.append(x_next)

        # Horizontal line: move from curve (x, f(x)) to diagonal (f(x), f(x))
        px.append(x_next)
        py.append(x_next)

        x = x_next

    return px, py

def plot_cobweb(ax, lam, x0, num_random_orbits=10, n_iter=30):
    """
    Draws a cobweb plot with one primary orbit and multiple random background orbits.
    """

    # Set plot limits and labels
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_xlabel('$x_n$')
    ax.set_ylabel('$x_{n+1}$')
    ax.set_title(f'$\lambda = {lam}$')

    # --- 1. Draw the Background (Diagonal and Parabola) ---

    # The y=x line
    x_diag = np.linspace(0, 1, 100)
    ax.plot(x_diag, x_diag, 'k--', linewidth=1, alpha=0.5)

    # The logistic function (parabola)
    x_func = np.linspace(0, 1, 400)
    ax.plot(x_func, logistic_map(x_func, lam), 'black', linewidth=1.5, alpha=0.8)

    # --- 2. Draw the Random Orbits (y) ---
    # We draw these first so they appear "behind" the main red orbit
    for _ in range(num_random_orbits):
        y_start = np.random.rand()
        bx, by = get_cobweb_path(lam, y_start, n_iter)
        # Low alpha (transparency) to handle overlap cleanly
        ax.plot(bx, by, 'b-', alpha=0.15, linewidth=0.8)

    # --- 3. Draw the Main Fixed Orbit (x0) ---
    rx, ry = get_cobweb_path(lam, x0, n_iter)
    ax.plot(rx, ry, 'r-', alpha=0.9, linewidth=1.5, label=f'$x_0={x0}$')

    # Add a dummy legend entry for the random orbits
    ax.plot([], [], 'b-', alpha=0.5, label=f'{num_random_orbits} random $y$')
    ax.legend(loc='upper left', fontsize='small', framealpha=0.9)

# --- Main execution ---
if __name__ == "__main__":
    # Setup a single figure with 4 subplots (2x2 grid)
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle("Logistic Map Cobweb Plots: Multiple Orbits", fontsize=16)

    axes_flat = axs.flatten()

    # Define scenarios
    scenarios = [
        {"lam": 2.8, "n_iter": 30, "desc": "Stable Fixed Point"},
        {"lam": 3.4, "n_iter": 50, "desc": "Period-2 Cycle"},
        {"lam": 3.5, "n_iter": 100, "desc": "Period-4 Cycle"},
        {"lam": 3.9, "n_iter": 200, "desc": "Chaos"}
    ]

    # Loop through scenarios
    for ax, params in zip(axes_flat, scenarios):
        x0 = 0.5
        # We pass 10 as the number of random orbits to generate
        plot_cobweb(ax, params["lam"], x0, num_random_orbits=10, n_iter=params["n_iter"])

        ax.text(0.05, 0.05, params["desc"], transform=ax.transAxes,
                fontsize=10, bbox=dict(facecolor='white', alpha=0.9))

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    print("Generating plots with 10 additional random y orbits per graph...")
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# --- Configuration Parameters ---
n_lambda_steps = 2000  # Number of different lambda values to plot on the x-axis
lambda_min = 1.0       # Minimum lambda value
lambda_max = 4.0       # Maximum lambda value

x0 = 0.5               # Initial condition for the iteration.
                       # This *must* be a value between 0 and 1, usually 0.5
                       # to capture the full dynamics of the logistic map.
                       # If x0=0, the iterates will always be 0, resulting in a blank diagram.

n_transient = 100      # Number of initial iterations to discard for each lambda.
                       # This allows the system to settle onto its attractor (fixed point, cycle, or chaotic behavior).
                       # Your request for n=100 suggests this starting point.

n_collect = 200        # Number of subsequent iterations to collect and plot for each lambda.
                       # Your request for n=100 to 200 means 101 points (n_collect = 101).
                       # Using more points (like 200 or 300) makes the chaotic regions look denser and better.

# --- Data Generation ---
print("Starting bifurcation diagram data generation...")
start_time = time.time()

# Create an array of lambda values from min to max, with `n_lambda_steps` points.
lambdas = np.linspace(lambda_min, lambda_max, n_lambda_steps)

# Lists to store the (lambda, x) pairs that will be plotted.
# We store them as separate lists to use with plt.plot efficiently.
lambda_plot_points = []
x_plot_points = []

# Iterate through each lambda value
for i, lam in enumerate(lambdas):
    # Print progress periodically to the console
    if (i + 1) % (n_lambda_steps // 10) == 0: # Print every 10% progress
        print(f"Progress: {i+1}/{n_lambda_steps} (lambda = {lam:.3f})")

    x = x0  # Reset x to the initial condition for each new lambda

    # Run `n_transient` iterations to let the system converge to its attractor
    for _ in range(n_transient):
        x = lam * x * (1 - x)

    # Collect `n_collect` subsequent iterations to plot
    for _ in range(n_collect):
        x = lam * x * (1 - x)

        # Append the current lambda and the current x value to our plotting lists
        lambda_plot_points.append(lam)
        x_plot_points.append(x)

print("...Data generation complete.")

# --- Plotting ---
print("Generating plot...")
plt.figure(figsize=(12, 8)) # Set a good figure size

# Plot all collected (lambda, x) points.
# 'k.' means black dots. markersize=0.02 makes them almost like pixels,
# which is essential for rendering a dense bifurcation diagram.
plt.plot(lambda_plot_points, x_plot_points, 'k.', markersize=0.02)

# Set title and labels
plt.title('Bifurcation Diagram of the Logistic Map', fontsize=18)
plt.xlabel(r'Parameter $\lambda$', fontsize=15) # Use r'...' for LaTeX rendering of lambda
plt.ylabel(r'Attractor Points $x_n$', fontsize=15)

# Set x and y axis limits
plt.xlim(lambda_min, lambda_max)
plt.ylim(0, 1) # Logistic map iterates always stay between 0 and 1

# Add a grid for better readability (optional for this plot type, often omitted)
plt.grid(False)

# Save the figure to a high-resolution PNG file
plt.savefig('bifurcation_diagram.png', dpi=300)

end_time = time.time()
print(f"Figure saved as 'bifurcation_diagram.png'.")
print(f"Total time taken: {end_time - start_time:.2f} seconds.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

def plot_bifurcation_zoom(lambda_min, lambda_max, y_min, y_max,
                           n_lambda_steps, n_transient, n_collect,
                           markersize, dpi, filename):
    """
    Generates and saves a zoomed-in bifurcation diagram for the logistic map.

    Args:
        lambda_min (float): Minimum lambda value (x-axis min).
        lambda_max (float): Maximum lambda value (x-axis max).
        y_min (float): Minimum iterate value (y-axis min).
        y_max (float): Maximum iterate value (y-axis max).

        n_lambda_steps (int): Horizontal resolution. Increase for narrower lambda_ranges.
        n_transient (int): Iterations to discard. Increase for deeper zooms.
        n_collect (int): Iterations to plot. Increase for denser plots.

        markersize (float): Plot marker size. Decrease for deeper zooms.
        dpi (int): Resolution of the saved image.
        filename (str): Output file name.
    """

    print(f"--- Starting Plot Generation ---")
    print(f"  Window: lambda=[{lambda_min}, {lambda_max}], y=[{y_min}, {y_max}]")
    print(f"  Params: steps={n_lambda_steps}, transient={n_transient}, collect={n_collect}")

    start_time = time.time()

    x0 = 0.5  # Standard initial condition

    # Create the array of lambda values
    lambdas = np.linspace(lambda_min, lambda_max, n_lambda_steps)

    # Lists to store the points for plotting
    lambda_plot_points = []
    x_plot_points = []

    # Main loop over each lambda value
    for i, lam in enumerate(lambdas):
        x = x0

        # 1. Run through the transient iterations (to let the system settle)
        for _ in range(n_transient):
            x = lam * x * (1 - x)

        # 2. Collect the iterates to be plotted
        for _ in range(n_collect):
            x = lam * x * (1 - x)

            # Optimization: Only store points that are within our y-axis zoom
            if y_min <= x <= y_max:
                lambda_plot_points.append(lam)
                x_plot_points.append(x)

    print(f"...Data generation complete ({len(lambda_plot_points)} points).")

    # 3. Plotting
    print("Generating plot...")
    plt.figure(figsize=(12, 8))

    # Use 'k.' for black pixels/dots
    plt.plot(lambda_plot_points, x_plot_points, 'k.', markersize=markersize)

    plt.title('Bifurcation Diagram (Zoomed)', fontsize=18)
    plt.xlabel(r'Parameter $\lambda$', fontsize=15)
    plt.ylabel(r'Attractor Points $x_n$', fontsize=15)

    # Apply the user-defined zoom window
    plt.xlim(lambda_min, lambda_max)
    plt.ylim(y_min, y_max)

    plt.savefig(filename, dpi=dpi)

    end_time = time.time()
    print(f"Figure saved as '{filename}'.")
    print(f"Total time taken: {end_time - start_time:.2f} seconds.")
    print("-" * 30)

# ---
# --- HOW TO USE ---
# ---
#
# You can now generate any view by calling the function.
# Below are two examples: the full diagram and our last deep zoom.
#

# --- Example 1: The "Deeper Zoom" from our last conversation ---
print("--- Example 1: Deeper Zoom (3-Cycle Window) ---")
plot_bifurcation_zoom(
    lambda_min = 3.8475,      # X-axis start
    lambda_max = 3.8495,      # X-axis end
    y_min = 0.490,            # Y-axis start
    y_max = 0.505,            # Y-axis end
    n_lambda_steps = 8000,    # High horizontal resolution
    n_transient = 1000,       # Long settling time
    n_collect = 800,          # High density
    markersize = 0.01,        # Tiny markers
    dpi = 600,                # High-res image
    filename = 'bifurcation_deeper_zoom.png'
)


# --- Example 2: The Full Diagram ---
print("--- Example 2: Full Diagram ---")
plot_bifurcation_zoom(
    lambda_min = 0.0,
    lambda_max = 4.0,
    y_min = 0.0,
    y_max = 1.0,
    n_lambda_steps = 2000,    # Standard resolution
    n_transient = 100,        # Standard settling time
    n_collect = 200,          # Standard density
    markersize = 0.02,
    dpi = 300,
    filename = 'bifurcation_full.png'
)
```
